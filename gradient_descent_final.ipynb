```python
from sklearn import preprocessing
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from sklearn import datasets
```

```python
data_train = np.loadtxt("training_boston.csv", delimiter=",")
data_test = np.loadtxt("test_boston.csv", delimiter=",")
```

```python
x_train=pd.DataFrame(data_train[:,:-1])
y_train=data_train[:,-1:]
x_test=pd.DataFrame(data_test.copy())
```

```python
df = pd.DataFrame(x_train)
df

for i in df.columns:
    name = str(i)+"_"+str(i)
    df[name] = df[i]**2
df.describe()
x2_train=df.values
x2_train.shape
```

```python
df1 = pd.DataFrame(x_test)
for i in df1.columns:
    name = str(i)+"_"+str(i)
    df1[name] = df1[i]**2
df1.describe()
x2_test = df1.values
x2_test.shape
```

```python
scaler = preprocessing.StandardScaler()
scaler.fit(x2_train)
x2_train = scaler.transform(x2_train)
x2_test = scaler.transform(x2_test)
```

```python
P = np.ones((len(x2_train[:, 0]), len(x2_train[0]) + 1))
P[:,:-1] = x2_train
x3_train = P

Q = np.ones((len(x2_test[:, 0]), len(x2_test[0]) + 1))
Q[:,:-1] = x2_test
x3_test = Q
```

```python
print(len(x3_train[0]),len(x3_train[:,0]))
x3_train[0][0]
```

```python
x3_train=np.array(x3_train)
x3_test=np.array(x3_test)
```

```python
def step_gradient(learning_rate,m,x,y):
    m_slope=np.zeros(x.shape[1])
    M=x.shape[0]
    for j in range(x.shape[1]):
        for i in range(M):
            m_slope[j]+=(-2/M)*(y[i]-(m*x[i]).sum())*x[i][j]
    new_m=m-(learning_rate*m_slope)
    return(new_m)
```

```python
def cost(x,y,m):
    M=x.shape[0]
    total_cost=0
    for i in range(M):
        total_cost+=(1/M)*((y[i]-(m*x[i]).sum())**2)
    return total_cost
```

```python
def gd(x,y,learning_rate,num_iterations):
    M = len(x[:,0])
    N = len(x[0])
    m = np.zeros(N)
    for i in range(num_iterations):
        m=step_gradient(learning_rate,m,x,y)
        print(i,"cost:",cost(x,y,m))
    
    return m

```

```python
def predict(x_test,m):
    M = len(x_test[:,0])
    Y_pred=np.zeros(M)
    for i in range(M):
        Y_pred[i]=(m*x_test[i]).sum()
    return Y_pred
```

```python
learning_rate = 0.064
num_of_iterations =500
m=gd(x3_train,y_train, learning_rate, num_of_iterations)
y_pred = predict(x3_test,m)
np.savetxt("boston_predict.csv",y_pred,"%.8f",delimiter=",")
```
